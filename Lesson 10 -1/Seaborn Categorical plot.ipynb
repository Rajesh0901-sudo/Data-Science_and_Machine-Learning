```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_dataset = sns.load_dataset('tips')
```

```python
tips_dataset
```

```python
print(tips_dataset.columns)
```

```python
sns.catplot(data=tips_dataset, x='day' , y='total_bill')
```

```python
sns.catplot(data=tips_dataset, x='day' , y='total_bill',kind='violin')
```

```python
sns.catplot(data=tips_dataset, x='day' ,kind='count')
```

```python
sns.catplot(data=tips_dataset, x='sex' ,kind='count',col='day',col_wrap=2)
```

```python
sns.catplot(data=tips_dataset, x='sex' ,kind='count',col='day',row='size')
```

```python
sns.catplot(data=tips_dataset, y='sex',x='total_bill' ,kind='point',col='day',col_wrap=2)
```
