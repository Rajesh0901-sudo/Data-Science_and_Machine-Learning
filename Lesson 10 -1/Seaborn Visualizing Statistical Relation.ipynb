```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
```

```python
dataset = sns.load_dataset('flights')
```

```python
print(dataset)
```

```python
style.use('ggplot')
plt.figure(figsize=(6,6))
sns.relplot(data=dataset , x='passengers', y='month')
```

```python

sns.relplot(data=dataset , x='passengers', y='month',hue='year')
```

```python
sns.relplot(data=dataset , x='passengers', y='year',row='month')
```

```python
flights_wide = sns.load_dataset("flights").pivot("year", "month", "passengers")
```

```python
sns.relplot(data=flights_wide,kind='line')
```
