#### 1: List the causes of death and percentile in Ohio state in the year 2012

```python
#import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
# View the plot 
plt.plot()
```

```python
#cause of deaths
Cause_of_death = ['Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis','Others']
#percentile
Percentile = np.array([62, 5, 4, 2, 1 , 26])
```

#### 2: Draw a pie chart using the two data points

```python
#set the pie chart plot properties
x = Percentile
y = Cause_of_death
plt.pie(x,labels=y)
```

```python
#set figure size
plt.figure(figsize=(9,9))
plt.pie(x,labels=y)
plt.show()
```

```python
#explode the largest pie in the data set
plt.figure(figsize=(9,9))
exp=[0.05,0,0,0,0,0]
plt.pie(x,labels=y,explode=exp)
plt.show()
```

```python
#set pie chart properties
plt.figure(figsize=(9,9))
exp=[0.05,0,0,0,0,0]
plt.pie(x,labels=y,explode=exp,shadow=True)
plt.legend()
plt.show()
```

```python
#set axis equal to draw pie as circle
plt.figure(figsize=(9,9))
exp=[0.05,0,0,0,0,0]
plt.pie(x,labels=y,explode=exp,shadow=True)
plt.legend()
plt.axis('equal')
plt.show()
```

```python
#set title of the pie chart
#set axis equal to draw pie as circle
plt.figure(figsize=(9,9))
exp=[0.05,0,0,0,0,0]
plt.pie(x,labels=y,explode=exp,shadow=True)
plt.legend()
plt.axis('equal')
plt.title('Death by Diseases')
```

```python
#show the plot (pie chart)

plt.figure(figsize=(9,9))
exp=[0.05,0,0,0,0,0]
plt.pie(x,labels=y,explode=exp,shadow=True)
plt.legend()
plt.axis('equal')
plt.title('Death by Diseases')
plt.show()
```
