```python
#importing libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('/Users/rajeshsahu/Desktop/Data Science Learning/Other/SpotifyFeatures 2.csv')
```

```python
dataset.head(10)
```

```python
duration = dataset.duration_ms.apply(lambda x:round(x/1000))
```

```python
plt.figure(figsize=(9,10))
plt.title('Duration of the songs ion different Geners')
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre',x=duration,data=dataset)
plt.xlabel('Duration in Seconds')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.set_style(style='whitegrid')
plt.title('Duration of the songs ion different Geners')
sns.color_palette("rocket",as_cmap=True)
sns.lineplot(x='genre',y=duration,data=dataset)
plt.xlabel('Duration in Seconds')
plt.ylabel('Genre')
plt.xticks(rotation=90)
plt.show()
```

```python
corr_df = dataset.corr(method='pearson')
```

```python
plt.figure(figsize=(9,6))
sns.heatmap(corr_df,annot=True,fmt='0.1g',linecolor='Black',linewidths=1)
```

```python
famous = dataset.sort_values('popularity',ascending=False).head(10)
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,6))
plt.title('Top Geners')
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre',x='popularity',data=famous)
plt.xlabel('Popularity')
plt.ylabel('Genre')
plt.show()
```
